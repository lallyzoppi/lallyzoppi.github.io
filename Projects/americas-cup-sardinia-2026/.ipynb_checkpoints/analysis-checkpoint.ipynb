# 🌊 America’s Cup Cagliari 2026 Analytics

Educational sailing analytics project using simulated telemetry data.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/boat_track.csv')

df.head()
```

## ⚡ Boat Speed Analysis

The graph shows how the boat speed changes over time.  
We can observe a progressive increase in speed until peak values are reached.
This behavior is consistent with a foiling phase during racing conditions.

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

df = df.sort_values('time').reset_index(drop=True)

# smoothing (come nel plot statico)
df['speed_smooth'] = df['speed_knots'].rolling(
    window=3, min_periods=1
).mean()

fig, ax = plt.subplots(figsize=(13,6))

ax.set_title("Boat Speed Animation")
ax.set_xlabel("Time")
ax.set_ylabel("Speed (knots)")

ax.set_xlim(df['time'].min(), df['time'].max())
ax.set_ylim(df['speed_knots'].min()-1, df['speed_knots'].max()+1)

# 🔵 RAW line
raw_line, = ax.plot([], [], color='steelblue', alpha=0.3, label='Raw')

# 🔵 SMOOTH line (quella importante)
smooth_line, = ax.plot([], [], color='navy', linewidth=3, label='Smoothed')

# 🔴 punto corrente
point, = ax.plot([], [], 'ro', markersize=6)

ax.legend()

def update(i):
    x = df['time'].iloc[:i+1]

    # RAW
    raw_line.set_data(x, df['speed_knots'].iloc[:i+1])

    # SMOOTH
    smooth_line.set_data(x, df['speed_smooth'].iloc[:i+1])

    # POINT
    point.set_data([df['time'].iloc[i]], [df['speed_knots'].iloc[i]])

    return raw_line, smooth_line, point

anim = FuncAnimation(
    fig,
    update,
    frames=len(df),
    interval=200,
    blit=False
)

plt.close(fig)

HTML(anim.to_jshtml())
```

## 🗺️ GPS Track Visualization

The route shows a curved trajectory typical of buoy racing courses.
We can observe several directional changes corresponding to tacks and gybes.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-darkgrid')

fig, ax = plt.subplots(figsize=(8,8))

# indice per creare colore progressivo (effetto traiettoria nel tempo)
t = np.arange(len(df))

# linea base
ax.plot(df['lon'], df['lat'], linewidth=2, color='steelblue', alpha=0.6)

# scatter con gradiente (traiettoria temporale)
sc = ax.scatter(
    df['lon'],
    df['lat'],
    c=t,
    cmap='viridis',
    s=25,
    edgecolor='k',
    linewidth=0.3
)

# punto di partenza
ax.scatter(
    df['lon'].iloc[0],
    df['lat'].iloc[0],
    color='green',
    s=120,
    label='Start',
    edgecolor='black'
)

# punto di arrivo
ax.scatter(
    df['lon'].iloc[-1],
    df['lat'].iloc[-1],
    color='red',
    s=120,
    label='End',
    edgecolor='black'
)

# styling
ax.set_title('Boat GPS Track', fontsize=16, weight='bold')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

ax.grid(True, alpha=0.3)
ax.legend()

# colorbar (tempo)
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Track Progress (time)')

plt.tight_layout()
plt.show()
```

## 🌬️ Wind vs Speed

```python
The scatter plot shows a positive relationship between wind speed and boat speed.
However, the relationship is not perfectly linear: after a certain threshold, performance tends to stabilize.
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-darkgrid')

fig, ax = plt.subplots(figsize=(9,6))

# scatter con colore progressivo (tempo)
t = np.arange(len(df))

sc = ax.scatter(
    df['wind_speed'],
    df['speed_knots'],
    c=t,
    cmap='viridis',
    s=60,
    alpha=0.8,
    edgecolor='black',
    linewidth=0.3
)

# trend line (regressione semplice)
z = np.polyfit(df['wind_speed'], df['speed_knots'], 1)
p = np.poly1d(z)

x_line = np.linspace(df['wind_speed'].min(), df['wind_speed'].max(), 100)
ax.plot(x_line, p(x_line), color='red', linewidth=2, label='Trend line')

# styling
ax.set_title('Wind Speed vs Boat Speed', fontsize=16, weight='bold')
ax.set_xlabel('Wind Speed (knots)')
ax.set_ylabel('Boat Speed (knots)')

ax.grid(True, alpha=0.3)
ax.legend()

# colorbar tempo
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Time Progress')

plt.tight_layout()
plt.show()
```
